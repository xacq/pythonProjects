```python
#Importación de módulos
import pandas as pd
import copy
import numpy as np
print('Módulos importados')
```

```python
#Carga del dataset
dfOriginal = pd.read_csv("master_dataset.csv", low_memory=False, sep=',', index_col=False, nrows=10000)
#Se reemplaza los NAN por el valor más frecuente (la moda) en cada columna (es decir, en cada variable)
dfOriginal = dfOriginal.fillna(dfOriginal.mode().iloc[0])
print(dfOriginal.shape)
dfOriginal.head(30)
```

```python
#Se elimina algunas variables
dataframe=copy.deepcopy(dfOriginal)
print(dataframe.shape)
dataframe=dataframe.drop(['survey_date'], axis=1)
dataframe=dataframe.drop(['region'], axis=1)
dataframe=dataframe.drop(['ip_latitude'], axis=1)
dataframe=dataframe.drop(['ip_longitude'], axis=1)
dataframe=dataframe.drop(['ip_accuracy'], axis=1)
dataframe=dataframe.drop(['working'], axis=1)
dataframe=dataframe.drop(['covid19_symptoms'], axis=1)
dataframe=dataframe.drop(['contacts_count'], axis=1)
dataframe=dataframe.drop(['house_count'], axis=1)
dataframe=dataframe.drop(['public_transport_count'], axis=1)
dataframe=dataframe.drop(['worried'], axis=1)
dataframe=dataframe.drop(['rate_reducing_mask_type'], axis=1)
dataframe=dataframe.drop(['health_worker'], axis=1)
dataframe=dataframe.drop(['prescription_medication'], axis=1)
dataframe=dataframe.drop(['opinion_infection'], axis=1)
dataframe=dataframe.drop(['opinion_mortality'], axis=1)
dataframe=dataframe.drop(['risk_infection'], axis=1)
dataframe['height'] = dataframe['height'].replace('undefined', 
    np.mean(pd.to_numeric(dataframe['height'], errors='coerce')))
dataframe['height'] = dataframe['height'].astype(float)
dataframe['weight'] = dataframe['weight'].replace('undefined', 
    np.mean(pd.to_numeric(dataframe['weight'], errors='coerce')))
dataframe['weight'] = dataframe['weight'].astype(float)
print(dataframe.shape)
dataframe.head(30)
```
