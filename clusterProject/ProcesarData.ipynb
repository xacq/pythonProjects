Lectura del archivo con los datos de la encuesta

```python
# Lee el archivo CSV
df = pd.read_csv('dataEncuesta.csv')
```

```python
# Muestra las primeras filas del DataFrame
print(df.head())
```

```python
# Function to replace individual age values with specified ranges
def replace_age(value):
    try:
        age = int(value)
        if 18 <= age <= 24:
            return "18 a 24 años"
        elif 55 <= age <= 64:
            return "55 a 64 años"
        elif age >= 65:
            return "Mayor a 65 años"
        else:
            return value
    except ValueError:
        return value

# Apply the function to the 'Edad' column
df['1.- Datos demográficos del cliente\n¿Qué edad tiene?:'] = df['1.- Datos demográficos del cliente\n¿Qué edad tiene?:'].apply(replace_age)

# Verify the changes
df['1.- Datos demográficos del cliente\n¿Qué edad tiene?:'].unique()
```

```python
# List of valid age ranges
valid_age_ranges = ['25 a 35 años', '35 a 44 años', '45 a 54 años', '55 a 64 años', 'Mayor a 65 años']

# Replace any other value with '45 a 54 años'
df['1.- Datos demográficos del cliente\n¿Qué edad tiene?:'] = df['1.- Datos demográficos del cliente\n¿Qué edad tiene?:'].apply(
    lambda x: x if x in valid_age_ranges else '45 a 54 años'
)

# Verify the changes
print(df['1.- Datos demográficos del cliente\n¿Qué edad tiene?:'].unique())
```

```python
# List of valid gender values
valid_genders = ['Masculino', 'Femenino']

# Replace any other value or missing value with 'Masculino'
df['2.- Género:'] = df['2.- Género:'].apply(lambda x: x if x in valid_genders else 'Masculino')

# Verify the changes
print(df['2.- Género:'].unique())
```

```python
# Replace NaN or missing values in the 'Nivel Educativo' column with 'Bachillerato'
df['3.- ¿Qué nivel educativo tiene?'] = df['3.- ¿Qué nivel educativo tiene?'].fillna('Bachillerato')

# Verify the changes
print(df['3.- ¿Qué nivel educativo tiene?'].unique())
```

```python
# Replace NaN or missing values in the 'Miembros de familia' column with '1-3 miembros'
df['4.- ¿Cuántos miembros compone su familia?'] = df['4.- ¿Cuántos miembros compone su familia?'].fillna('1-3 miembros').replace(['', 'nan', 'NaN'], '1-3 miembros')

# Verify the changes
print(df['4.- ¿Cuántos miembros compone su familia?'].unique())
```

```python
# Replace NaN or missing values in the 'Frecuencia de conduccion' column with 'Varias veces a la semana'
df['5.- ¿Con qué frecuencia conduce?'] = df['5.- ¿Con qué frecuencia conduce?'].fillna('Varias veces a la semana').replace(['', 'nan', 'NaN'], 'Varias veces a la semana')

# Verify the changes
print(df['5.- ¿Con qué frecuencia conduce?'].unique())
```

```python
# List of valid frequency values
valid_frequencies = ['Diariamente', 'Varias veces a la semana', 'Una vez a la semana', 'Menos de una vez a la semana']

# Replace any other value with 'Nunca'
df['5.- ¿Con qué frecuencia conduce?'] = df['5.- ¿Con qué frecuencia conduce?'].apply(lambda x: x if x in valid_frequencies else 'Nunca')

# Verify the changes
print(df['5.- ¿Con qué frecuencia conduce?'].unique())
```
